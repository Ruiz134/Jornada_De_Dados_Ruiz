## Gabarito — Exercício 50

```python
cadastro = {}

for i in range(1, 4):
    nome = input(f'Digite o nome da pessoa {i}: ')
    idade = int(input(f'Digite a idade da pessoa {i}: '))
    cidade = input(f'Digite a cidade da pessoa {i}: ')

    cadastro[f'pessoa_{i}'] = {
        'nome': nome,
        'idade': idade,
        'cidade': cidade
    }

for chave, dados in cadastro.items():
    print(chave, dados)
```
